```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import wbgapi as wb
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# indicator id
indicator_id = ['EG.ELC.RNEW.ZS','EG.FEC.RNEW.ZS']

# Country code
country_code = ['AUS','CHN','CAN','FIN','IND','MDV','RUS','NZL','DEU','USA','ARG']

# Country name
country_name = {'AUS':'Australia','CHN':'China','CAN':'Canada','FIN':'Finland','IND':'India',
                'MDV':'Maldives','RUS':'Russia','NZL':'New Zealand','DEU':'Germany','USA':'United States',
                'ARG':'Argentina'}
```

```python
wb.series.info(indicator_id)
```

```python
# Functions
def readData(file_id,country_codes):
    file = pd.DataFrame(wb.data.DataFrame(file_id, country_codes, time=range(2000, 2015)))
    file_tr = file.transpose()
    return file, file_tr
```

```python
# Read Data and its transpose
ele_out, ele_out_trans = readData('EG.ELC.RNEW.ZS',country_code)
ele_con, ele_con_trans = readData('EG.FEC.RNEW.ZS',country_code)
```

#### Mean and variance in electricty production

```python
mean_out = {}
var_out = {}
for i in ele_out_trans:
    mean_out[i] = ele_out_trans[i].mean()
    var_out[i] = ele_out_trans[i].var()

#Total mean and variance
mean_tot_out = ele_out_trans.stack().mean()
var_tot_out = ele_out_trans.stack().var()

five_mean_out = sorted(mean_out, key = mean_out.get, reverse = True)[:5]
five_var_out = sorted(var_out, key = var_out.get, reverse = True)[:5]
```

```python
# Total mean of whole dataset
print(f'The total mean of whole data is {round(mean_tot_out,2)}')
print('')

# Top 5 countries
print(f'The top 5 countries with highest mean production of electricity are')
for i in range(5):
    print(f'{country_name[five_mean_out[i]]} with {round(mean_out[five_mean_out[i]],2)}')
```

```python
# Total variance of whole dataset
print(f'The total variance of whole data is {round(var_tot_out,2)}')
print('')

# Top 5 countries
print(f'The top 5 countries with highest variance in production of electricity are')
for i in range(5):
    print(f'{country_name[five_var_out[i]]} with {round(var_out[five_var_out[i]],2)}')

```

#### Mean and variance of electricty consumtion

```python
mean_con = {}
var_con = {}
for i in ele_con_trans:
    mean_con[i] = ele_con_trans[i].mean()
    var_con[i] = ele_con_trans[i].var()

#Total mean and variance
mean_tot_con = ele_con_trans.stack().mean()
var_tot_con = ele_con_trans.stack().var()

five_mean_con = sorted(mean_con, key = mean_con.get, reverse = True)[:5]
five_var_con = sorted(var_con, key = var_con.get, reverse = True)[:5]

```

```python
# Total mean of whole dataset
print(f'The total mean of whole data is {round(mean_tot_con,2)}')
print('')

# Top 5 countries
print(f'The top 5 countries with highest mean consumtion of electricity are')
for i in range(5):
    print(f'{country_name[five_mean_con[i]]} with {round(mean_con[five_mean_con[i]],2)}')

```

```python
# Total variance of whole dataset
print(f'The total variance of whole data is {round(var_tot_con,2)}')
print('')

# Top 5 countries
print(f'The top 5 countries with highest variance in consumtion of electricity are')
for i in range(5):
    print(f'{country_name[five_var_con[i]]} with {round(var_con[five_var_con[i]],2)}')


```

### Standerd deveation

```python
std_out = ele_out.std(axis=1)
std_con = ele_con.std(axis=1)

plt.plot(std_out)
plt.plot(std_con)
plt.title('Standerd deveation of electricity of each country')
plt.xlabel('Country')
plt.ylabel('std div')
plt.savefig('std_country.png', bbox_inches='tight')
plt.show()
```

### Correlataion

```python
corr_dist = {}
for i in country_code:
    corr_dist[i] = np.corrcoef(ele_con.loc[i],ele_out.loc[i])[1][0]

corr_dist_sorted = sorted(corr_dist,key = corr_dist.get, reverse = True)
lists = sorted(corr_dist.items())
x, y = zip(*lists)
plt.bar(x,y)
plt.title('Correlation btw electric consumtion and production')
plt.ylabel('Correlation')
plt.xlabel('Countries')
plt.savefig('corr_country', bbox_inches='tight')
plt.show()
```

```python
# Correlation between production and cunsumtion
print('The countries with highest positive correlation are :')
for i in corr_dist_sorted:
    if corr_dist[i] >= 0.5:
        print(f'{country_name[i]} : {round(corr_dist[i],2)}')
print(' ')       
print('The countries with highest negative correlation are :')
for i in corr_dist_sorted:
    if corr_dist[i] <= -0.5:
        print(f'{country_name[i]} : {round(corr_dist[i],2)}')
print(' ')
print('The countries with low correlation are :')
for i in corr_dist_sorted:
    if corr_dist[i] >= -0.5 and corr_dist[i] <= 0.5:
        print(f'{country_name[i]} : {round(corr_dist[i],2)}')
```

```python
# full data correlation
arg_con = ele_con.stack()
arg_out = ele_out.stack()
total_corr = np.corrcoef(arg_con,arg_out)[0][1]
print(f'The correlation between electric porduction and consumtion in the whole world is {round(total_corr,2)}')
```

```python
# Time series
time_corr = {}
for i in country_code:
    con_cor = ele_con.loc[i].pct_change()
    out_cor = ele_out.loc[i].pct_change() 
    x = con_cor.corr(out_cor).tolist()
    if x > 0 :
        time_corr[i] = x
    else: 
        time_corr[i] = 0

print('Time series correlation of each country is ')
for i in time_corr:
    print(f'{country_name[i]} : {round(time_corr[i],2)}')
```

```python
# time series data visualisation
plt.scatter(time_corr.keys(),time_corr.values())
plt.title('Time series correlation btw electric consumtion and production')
plt.ylabel('Correlation')
plt.xlabel('Countries')
plt.savefig('time_series_corr.png',bbox_inches='tight')
plt.show()
```
